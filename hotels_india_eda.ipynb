```python
# Group by 'State' and sum the 'Total Rooms'
state_rooms = df.groupby('State')['Total Rooms'].sum().reset_index()

# Sort based on the total rooms in descending order
state_rooms = state_rooms.sort_values(by='Total Rooms', ascending=False)

# Plot the top 10 states with the most total rooms
top_10_states = state_rooms.head(10)

plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_states['State'], top_10_states['Total Rooms'], color='salmon')
plt.title('Top 10 States with Most Total Rooms')
plt.xlabel('State')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Total Rooms')

# Display the values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.show()
```

```python
# Set the figure size using rcParams
plt.rcParams['figure.figsize'] = (30, 15)

# Assuming s_rooms is your DataFrame with State and Total Rooms columns
# Set a 'rainbow' color palette for the bars
colors = sns.color_palette('rainbow', len(state_rooms))

p = sns.barplot(x='State', y='Total Rooms', data=state_rooms, palette=colors)

# Add labels to the bars
for bar, label in zip(p.patches, state_rooms['Total Rooms']):
    p.annotate(label, (bar.get_x() + bar.get_width() / 2, bar.get_height()),
               ha='center', va='bottom')

plt.title("Number of Rooms in Each State")
plt.xticks(rotation=35, ha='right')
plt.xlabel('State')
plt.ylabel('Total Rooms')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Fill NaN values in 'Alcohol' column with 'Unknown'
df['Alcohol'].fillna('Unknown', inplace=True)

# Group by 'City' and count occurrences of each unique value in 'Alcohol'
city_alcohol_counts = df.groupby('City')['Alcohol'].value_counts().unstack(fill_value=0)

# Get top 10 cities for each category
top_10_with_alcohol = city_alcohol_counts['With Alcohol'].nlargest(10)
top_10_without_alcohol = city_alcohol_counts['Without Alcohol'].nlargest(10)
top_10_dont_know = city_alcohol_counts["dont know"].nlargest(10)

# Plot 'With Alcohol'
plt.figure(figsize=(12, 6))
plt.bar(top_10_with_alcohol.index, top_10_with_alcohol, label='With Alcohol', color='green')
plt.title('Top 10 Cities with Alcohol Availability')
plt.xlabel('City')
plt.ylabel('Count')
plt.legend()
plt.xticks(rotation=35, ha='right')
plt.show()

# Plot 'Without Alcohol'
plt.figure(figsize=(12, 6))
plt.bar(top_10_without_alcohol.index, top_10_without_alcohol, label='Without Alcohol', color='red')
plt.title('Top 10 Cities without Alcohol Availability')
plt.xlabel('City')
plt.ylabel('Count')
plt.legend()
plt.xticks(rotation=35, ha='right')
plt.show()

# Plot 'Don't Know'
plt.figure(figsize=(12, 6))
plt.bar(top_10_dont_know.index, top_10_dont_know, label="dont know", color='blue')
plt.title("Top 10 Cities with 'Don't Know' Alcohol Availability")
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=35, ha='right')
plt.legend()
plt.show()
```
