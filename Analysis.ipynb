```python

# Assuming df is your DataFrame with a 'Beverage_category' column

# Specify your desired colors
colors = sns.color_palette("pastel")  # You can choose other color palettes from seaborn

# Plotting the bar chart with specified colors
ax = df['Beverage_category'].value_counts().plot(kind='bar', color=colors)

# Rotate x-axis labels by 45 degrees
plt.xticks(rotation=45, ha='right')

# Annotate the top of each bar with its value
for i, v in enumerate(df['Beverage_category'].value_counts()):
    plt.text(i, v + 0.1, str(v), ha='center', va='bottom')

# Display the plot
plt.show()
```

```python
sns.pairplot(df, hue='Beverage_category', palette='husl', markers='D', diag_kind='kde', kind='reg')
```

```python
numeric_features = df.columns[df.dtypes != 'object'].tolist()
numeric_features
```

```python
def types_of_products(row):
    if row['Beverage_category'] == 'Tazo® Tea Drinks':
        val = 'Tea'
    elif row['Beverage_category'] == 'Smoothies':
        val = 'Smoothie'
    else:
        val = 'coffee'
    return val

df['Type_of_product'] = df.apply(types_of_products, axis=1)
```

```python
type_plot = sns.countplot(df, x="Type_of_product", order=df['Type_of_product'].value_counts(ascending=False).index)

type_plot.set(ylabel='N° of products', xlabel='Main categories', title = "Categories of Starbucks products")
sns.despine()
type_plot.bar_label(type_plot.containers[0])
```

```python
plt.figure(figsize=(14, len(numeric_features)*3))

for i, col in enumerate(numeric_features, 1):  # Start index from 1 for better subplot indexing
    plt.subplot(len(numeric_features)//2 + 1, 2, i)
    sns.kdeplot(data=df, x=col, hue="Type_of_product", palette="pastel", common_norm=False)  # Added common_norm parameter
    plt.title(col)
    plt.xlabel('')  # Clear x-axis label for better aesthetics
    plt.ylabel('Density')  # Assuming you want to label the y-axis as Density

plt.tight_layout()
plt.show()
```

```python
# Sort the DataFrame by calories for better visualization
sorted_df = df.sort_values(by='Calories', ascending=False)

# Assign a different color to each beverage using the viridis colormap
# colors = plt.cm.viridis(np.linspace(0, 1, len(sorted_df)))

colors = plt.cm.plasma(np.linspace(0, 1, len(sorted_df)))
# Create a bar chart with different colors for each beverage
plt.figure(figsize=(32, 18))
bars = plt.bar(sorted_df['Beverage'], sorted_df['Calories'], color=colors)

# Adding data labels on top of each bar
for bar, value in zip(bars, sorted_df['Calories']):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height() + 5, str(value), ha='center', color='black')

plt.xticks(rotation=45, ha='right')
plt.xlabel('Beverage')
plt.ylabel('Calories')
plt.title('Calories in Starbucks Beverages')
plt.tight_layout()
plt.show()
```
